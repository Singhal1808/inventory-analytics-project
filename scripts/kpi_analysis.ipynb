```python
import pandas as pd

df = pd.read_csv("../data/cleaned_sales_data.csv")
```

```python
# Revenue
df["Revenue"] = (
    df["sales"] *
    df["price"]
)
```

```python
# 7-day moving average demand forecast
df["Demand_Forecast"] = (
    df.groupby(["store_id", "item_id"])["sales"]
    .transform(lambda x: x.rolling(7).mean())
)
```

```python
# Estimated inventory
df["Inventory_Level"] = (
    df["sales"] * 1.5
)
```

```python
# Stockout risk
df["Stockout_Risk"] = (
    df["Demand_Forecast"] >
    df["Inventory_Level"]
).astype(int)
```

```python
# Sales growth
df["Sales_Growth"] = (
    df.groupby(["store_id", "item_id"])["sales"]
    .diff()
)
```

```python
df.to_csv("../data/kpi_sales_data.csv", index=False)
```
